1.Simulate exam scores for 500 students with mean = 70, std dev = 10.

Tasks: Probability of scoring above 80, below 50, between 60–75.

Visuals: Histogram with fitted  curve.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean_score, std_score, n_students = 70, 10, 500

scores = np.random.normal(mean_score, std_score, n_students)

prob_above_80 = np.mean(scores > 80)
prob_below_50 = np.mean(scores < 50)
prob_between_60_75 = np.mean((scores >= 60) & (scores <= 75))

print(f"Probability > 80: {prob_above_80:.3f}")
print(f"Probability < 50: {prob_below_50:.3f}")
print(f"Probability 60–75: {prob_between_60_75:.3f}")

plt.figure(figsize=(10,6))
plt.hist(scores, bins=20, density=True, alpha=0.6, color='skyblue', edgecolor='black')
x = np.linspace(min(scores), max(scores), 100)
plt.plot(x, norm.pdf(x, mean_score, std_score), 'r', lw=2)
plt.title("Exam Scores Histogram with Normal Curve")
plt.xlabel("Score")
plt.ylabel("Density")
plt.show()
```

2.Daily Temperature Prediction

Use past temperature data or simulate a data following normal trend

Find probabilities of extreme heat or cold.

Compare mean, median, and mode.
mean_temp = 30
std_temp = 5
days = 365


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, mode

mean_temp = 30
std_temp = 5
days = 365

temps = np.random.normal(mean_temp, std_temp, days)

extreme_heat = 40
extreme_cold = 20

prob_extreme_heat = np.mean(temps > extreme_heat)
prob_extreme_cold = np.mean(temps < extreme_cold)

mean_val = np.mean(temps)
median_val = np.median(temps)
mode_val = float(mode(np.round(temps, 1)).mode)

print(f"Probability of extreme heat (> {extreme_heat}°C): {prob_extreme_heat:.3f}")
print(f"Probability of extreme cold (< {extreme_cold}°C): {prob_extreme_cold:.3f}")
print(f"Mean temperature: {mean_val:.2f}°C")
print(f"Median temperature: {median_val:.2f}°C")
print(f"Mode temperature: {mode_val:.2f}°C")

plt.figure(figsize=(10,6))
plt.hist(temps, bins=20, density=True, alpha=0.6, color='orange', edgecolor='black')
x = np.linspace(min(temps), max(temps), 100)
plt.plot(x, norm.pdf(x, mean_temp, std_temp), 'b', lw=2)
plt.title("Daily Temperatures Histogram with Normal Curve")
plt.xlabel("Temperature (°C)")
plt.ylabel("Density")
plt.show()

```

3.Customer First Purchase Day

Simulate the day on which a new website visitor makes their first purchase (probability per day = 0.1).

Find probability that a customer buys on Day 1, Day 3, etc.

Visualize with PMF.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p = 0.1
max_days = 20

pmf_values = geom.pmf(np.arange(1, max_days+1), p)

for day, prob in enumerate(pmf_values, 1):
    print(f"Day {day}: {prob:.3f}")

plt.bar(np.arange(1, max_days+1), pmf_values, color='skyblue', edgecolor='black')
plt.show()
```

4.Basketball Free Throw Attempts

Probability of making a free throw = 0.6.

Find the probability distribution of the number of attempts until the first basket.

Discuss expectation and variance.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import geom

p = 0.6
max_attempts = 10

pmf_values = geom.pmf(np.arange(1, max_attempts+1), p)
expectation = geom.mean(p)
variance = geom.var(p)

for attempt, prob in enumerate(pmf_values, 1):
    print(f"Attempt {attempt}: {prob:.3f}")

print(f"Expected attempts: {expectation:.2f}")
print(f"Variance: {variance:.2f}")

plt.bar(np.arange(1, max_attempts+1), pmf_values, color='lightgreen', edgecolor='black')
plt.title("PMF of Attempts Until First Basket")
plt.xlabel("Attempt Number")
plt.ylabel("Probability")
plt.show()
```

7.Bus Arrival Time Simulation

If a bus can arrive anytime between 0–20 minutes after schedule, model it as uniform distribution.

Find the probability the bus arrives in the first 5 minutes or last 5 minutes.

Visualize PDF.
Waiting Time Between Customer Arrivals

If customers arrive every 6 minutes on average, model waiting times using exponential distribution.

Find the probability that waiting time exceeds 10 minutes.

Visualize CDF.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, expon

a, b = 0, 20
bus_times = uniform(a, b-a)

prob_first5 = bus_times.cdf(5)
prob_last5 = 1 - bus_times.cdf(15)

print(f"Probability bus arrives in first 5 minutes: {prob_first5:.3f}")
print(f"Probability bus arrives in last 5 minutes: {prob_last5:.3f}")

x = np.linspace(a, b, 100)
plt.plot(x, bus_times.pdf(x), 'b', lw=2)
plt.fill_between(x, 0, bus_times.pdf(x), where=(x<=5)|(x>=15), color='orange', alpha=0.5)
plt.title("PDF of Bus Arrival Times")
plt.xlabel("Minutes After Schedule")
plt.ylabel("Density")
plt.show()
```
