```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. create a line chart to denote the mean price per item.  

```python
df=pd.read_csv("bigmart_data.csv")
df.head(5)
```

```python
meanprice = df.groupby('Item_Identifier')['Item_MRP'].mean().reset_index().head(10)
print(meanprice)
```

```python
sns.lineplot(x='Item_Identifier', y='Item_MRP', data=meanprice, marker='o')
```

```python
meansales = df.groupby('Outlet_Type')['Item_Outlet_Sales'].mean().reset_index()
print(meanprice)
```

2.What is the mean sales for each outlet type? (both bar chart, pie chart )

# Bar Plot

```python
plt.figure(figsize=(8,5))
sns.barplot(x='Outlet_Type', y='Item_Outlet_Sales', data=meansales, palette='viridis')

plt.title("Mean Sales per Outlet Type (Bar Chart)")
plt.xlabel("Outlet Type")
plt.ylabel("Mean Sales")
plt.xticks(rotation=30)
```

# Pie PLot

```python
plt.figure(figsize=(6,6))
plt.pie(meansales['Item_Outlet_Sales'], 
        labels=meansales['Outlet_Type'], 
        autopct='%1.1f%%')

plt.title("Mean Sales per Outlet Type (Pie Chart)")
```

3.Histogram of Item_MRP using both sns and plt

# using matplotlib

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Item_MRP'], bins=20, color='green', edgecolor='black')

plt.title("Histogram of Item MRP (Matplotlib)")
plt.xlabel("Item MRP")
plt.ylabel("Frequency")

```

# using seaborn

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Item_MRP'], bins=20, kde=True, color='blue')

plt.title("Histogram of Item MRP (Seaborn)")
plt.xlabel("Item MRP")
plt.ylabel("Frequency")
```

4.Box plot of item weight

```python
plt.figure(figsize=(6,5))
plt.boxplot(df['Item_Weight'].dropna()) 
plt.title("Box Plot of Item Weight (Matplotlib)")
plt.ylabel("Item Weight")
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(y='Item_Weight', data=df)   # y-axis for vertical boxplot
plt.title("Box Plot of Item Weight (Seaborn)")
plt.ylabel("Item Weight")
```

5.Relative distribution of item weight and its visibility(scatter plot)

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Item_Weight'], df['Item_Visibility'], alpha=0.5, color='blue')

plt.title("Scatter Plot of Item Weight vs Item Visibility (Matplotlib)")
plt.xlabel("Item Weight")
plt.ylabel("Item Visibility")
plt.grid(True)
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='Item_Weight', y='Item_Visibility', data=df, alpha=0.5, color='green')

plt.title("Scatter Plot of Item Weight vs Item Visibility (Seaborn)")
plt.xlabel("Item Weight")
plt.ylabel("Item Visibility")
```

6.Violin plot for Item_Outlet_Sales

```python
plt.figure(figsize=(7,6))
sns.violinplot(y='Item_Outlet_Sales', data=df, color="skyblue")

plt.title("Violin Plot of Item Outlet Sales")
plt.ylabel("Item Outlet Sales")
```

```python
plt.figure(figsize=(7,6))
plt.violinplot(df['Item_Outlet_Sales'].dropna(), showmeans=True)

plt.title("Violin Plot of Item Outlet Sales (Matplotlib)")
plt.ylabel("Item Outlet Sales")
plt.show()
```

7.Scatter plot using seaborn and add hue

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(
    x='Item_Weight',
    y='Item_Visibility',
    data=df,
    hue='Item_Fat_Content',   # color points by fat content
    alpha=0.7
)

plt.title("Scatter Plot of Item Weight vs Item Visibility (with Hue)")
plt.xlabel("Item Weight")
plt.ylabel("Item Visibility")
```

8.Heat map for selected numeric columns


```python
numeric_df = df[['Item_Weight', 'Item_Visibility', 'Item_MRP', 'Item_Outlet_Sales']]


corr = numeric_df.corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Heatmap of Correlation Between Numeric Columns")
```
