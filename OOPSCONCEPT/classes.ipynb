```python
class BankAccount:
    def __init__(self, balance):
        self.__balance = balance   # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__add_interest()  # Call private method before deposit
            self.__balance += amount

    def __add_interest(self):
        self.__balance += self.__balance * 0.01 

    def get_balance(self):
        return self.__balance  # Public method to view balance


# Example usage
account = BankAccount(1000)  
account.deposit(500)        
print(account.get_balance()) 
```

```python
import datetime


class BankAccount:
    bank_name = "AES National Bank"  # Class attribute

    def __init__(self, account_number, owner, balance=0.0, account_type="Savings"):
        self.__account_number = account_number
        self.__owner = owner
        self.__balance = balance
        self.__account_type = account_type
        self.__transactions = []  # Store transaction history

    # ---------------- Private methods ----------------
    def __add_interest(self):
        if self.__account_type.lower() == "savings":
            interest = self.__balance * 0.02  # 2% interest
            self.__balance += interest
            self.__log_transaction("Interest", interest)

    def __log_transaction(self, transaction_type, amount):
        self.__transactions.append({
            "type": transaction_type,
            "amount": amount,
            "balance": self.__balance,
            "time": datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        })

    # ---------------- Public methods ----------------
    def deposit(self, amount):
        if amount > 0:
            self.__add_interest()
            self.__balance += amount
            self.__log_transaction("Deposit", amount)
        else:
            raise ValueError("Deposit amount must be positive.")

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError("Withdrawal amount must be positive.")
        if amount > self.__balance:
            raise ValueError("Insufficient balance.")
        self.__balance -= amount
        self.__log_transaction("Withdrawal", amount)

    def transfer(self, target_account, amount):
        if not isinstance(target_account, BankAccount):
            raise TypeError("Target account must be a BankAccount.")
        self.withdraw(amount)
        target_account.deposit(amount)
        self.__log_transaction("Transfer Out", amount)
        target_account.__log_transaction("Transfer In", amount)

    def get_balance(self):
        return self.__balance

    def get_account_info(self):
        return {
            "Account Number": self.__account_number,
            "Owner": self.__owner,
            "Type": self.__account_type,
            "Balance": self.__balance
        }

    def get_transaction_history(self):
        return self.__transactions

    # ---------------- Class & Static Methods ----------------
    @classmethod
    def change_bank_name(cls, new_name):
        cls.bank_name = new_name

    @staticmethod
    def validate_account_number(number):
        return str(number).isdigit() and len(str(number)) == 10


# ---------------- Menu-Driven Program ----------------
if __name__ == "__main__":
    # Create two accounts for demo
    acc1 = BankAccount("123", "Nirmal", 67000, "Savings")
    acc2 = BankAccount("987", "Vineetha", 347500, "Current")

    accounts = {
        "123": acc1,
        "987": acc2
    }

    print("üè¶ Welcome to", BankAccount.bank_name)

    while True:
        print("\n--- MENU ---")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Transfer")
        print("4. Check Balance")
        print("5. Account Info")
        print("6. Transaction History")
        print("7. Exit")

        choice = input("Enter your choice: ")

        try:
            if choice == "1":  # Deposit
                acc_no = input("Enter your account number: ")
                if acc_no in accounts:
                    amt = float(input("Enter amount to deposit: "))
                    accounts[acc_no].deposit(amt)
                    print("Deposit successful!")
                else:
                    print("Account not found.")

            elif choice == "2":  # Withdraw
                acc_no = input("Enter your account number: ")
                if acc_no in accounts:
                    amt = float(input("Enter amount to withdraw: "))
                    accounts[acc_no].withdraw(amt)
                    print("Withdrawal successful!")
                else:
                    print(" Account not found.")

            elif choice == "3":  # Transfer
                src = input("Enter your account number: ")
                dest = input("Enter target account number: ")
                if src in accounts and dest in accounts:
                    amt = float(input("Enter amount to transfer: "))
                    accounts[src].transfer(accounts[dest], amt)
                    print("Transfer successful!")
                else:
                    print(" One or both accounts not found.")

            elif choice == "4":  # Check Balance
                acc_no = input("Enter your account number: ")
                if acc_no in accounts:
                    print("Balance:", accounts[acc_no].get_balance())
                else:
                    print("Account not found.")

            elif choice == "5":  # Account Info
                acc_no = input("Enter your account number: ")
                if acc_no in accounts:
                    print(" Account Info:", accounts[acc_no].get_account_info())
                else:
                    print("Account not found.")

            elif choice == "6":  # Transaction History
                acc_no = input("Enter your account number: ")
                if acc_no in accounts:
                    print(" Transaction History:")
                    for t in accounts[acc_no].get_transaction_history():
                        print(t)
                else:
                    print(" Account not found.")

            elif choice == "7":  # Exit
                print(" Thank you for banking with", BankAccount.bank_name)
                break

            else:
                print("Invalid choice, try again.")

        except Exception as e:
            print("‚ö†Ô∏è Error:", e)
```
