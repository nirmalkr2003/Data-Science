```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
categories = ["A", "B", "C", "D"]
values = [3, 7, 5, 6]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

```python
x = [5, 7, 8, 7, 6, 9, 5, 6]
y = [99, 86, 87, 88, 100, 86, 103, 87]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.title("Histogram")
plt.show()
```

```python
sizes = [30, 25, 20, 15, 10]
labels = ["A", "B", "C", "D", "E"]

plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```
