```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 2, 1.5]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color="skyblue", edgecolor="black")
plt.title("Histogram")
plt.show()
```
