```python
plt.figure(figsize=(8,5))
dept_counts=df["Department"].value_counts()
sns.barplot(x=dept_counts.index,y=dept_counts.values)
plt.xlabel('Department')
plt.ylabel('Number of employees')
plt.title('Employees per Department')

```

```python
plt.figure(figsize=(7,5))
sns.histplot(df['Age'].dropna(), bins=10)   # or use sns.histplot(df['Age'])
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution (Histogram)')

plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(df['Salary'],df['Experience'],marker='o')
plt.xlabel('Joining Date')
plt.ylabel('Salary')
plt.title('Salary over Joining Date')
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(df['Salary'], df['Experience'])
plt.xlabel('Salary')
plt.ylabel('Performance Score')
plt.title('Salary vs Performance Score')
plt.grid(True)
```

```python
deptcount=df['Department'].value_counts()
plt.pie(deptcount.values,labels=deptcount.index,autopct='%1.1f%%',explode=[0, 0, 0] )
plt.title('Department-wise Employee Distribution')
# plt.axis('equal')
plt.legend(title='Departments', loc='upper right')
```

```python
plt.figure(figsize=(6,5))
sns.boxplot(x=df['Salary'])
plt.ylabel('Salary')
plt.title('Salary Boxplot')
```

```python
plt.figure(figsize=(6,5))
sns.violinplot(y=df['Salary'])
plt.ylabel('Salary')
plt.title('Salary Violin Plot')
```
