```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np

n, p = 1000, 0.05
x = np.arange(30, 70)
y = binom.pmf(x, n, p)

plt.bar(x, y, color='skyblue')
plt.bar(50, binom.pmf(50, n, p), color='orange')  # highlight X=50
plt.title('Binomial Distribution')
plt.xlabel('X (No. of Clicks)')
plt.ylabel('P(X)')
plt.show()

print(f"P(X=50): {binom.pmf(50, n, p)*100:.2f}%")
```
