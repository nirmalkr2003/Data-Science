```python
def fibi(n):
  if n<=1:
    return n
  else:
    return fibi(n-1)+fibi(n-2)
  
num=int(input())
for i in range(num):
  print(fibi(i),end=" ")
```

```python
def prime(n):
  for i in range(2,n):
    if n%i==0:
      print("Not a Prime")
      break
  else:
    print("Prime no")

n=int(input("enter"))
prime(n)
```
